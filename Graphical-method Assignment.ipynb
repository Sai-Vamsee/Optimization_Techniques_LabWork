```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,7,1000)

y1 = (24-6*x)/4
y2 = (6-x)/2
y3 = 1+x
y4 = 2+x-x
y5 = 0+x-x

a,b = np.meshgrid(x,x)

plt.plot(x,y1,c='blue',label="6x+4y<=24")
plt.plot(x,y2,c='orange',label="x+2y<=6")
plt.plot(x,y3,c='green',label="-x+y<=1")
plt.plot(x,y4,c='red',label="y<=2")
plt.plot(x,y5,c='purple',label="x>=0")
plt.xlim((0,10))
plt.ylim((0,10))    
plt.title(f'The Reddy Mikks Problem')

plt.imshow(((6*a+4*b<=24) & (a+2*b<=6) & (-a+b<=1) & (b<=2)) & (a>=0) & (b>=0).astype(int) , 
                extent=(a.min(),a.max(),b.min(),b.max()),origin="lower", cmap="Blues", alpha = 0.8);
plt.legend(loc = 1)

plt.show()
```
